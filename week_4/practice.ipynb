```python
import cv2 as cv
import numpy as np
```

## Sobel Edge Detection

```python
#Load the image
image = cv.imread("../images/NNEKA.png", cv.IMREAD_GRAYSCALE)

#resize the image
image = cv.resize(image, (500, 400))

#apply sobel edge detection
sobel_x = cv.Sobel(image, cv.CV_64F, 1, 0, ksize = 3)
sobel_y = cv.Sobel(image, cv.CV_64F, 0, 1, ksize = 3)

#combine the gradient images
sobel_combined = cv.magnitude(sobel_x, sobel_y)

#display the original image and the sobel edges
cv.imshow("Original", image)
cv.imshow("Sobel Edges", np.uint8(sobel_combined))

#wait for a key press and close all windows
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```

```python
#Load the image
image = cv.imread("../images/Sola-Oni-820x400-1-600x293.jpg", cv.IMREAD_GRAYSCALE)

#resize the image
image = cv.resize(image, (500, 400))

#apply prewitt edge detection
kernel_x = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
kernel_y = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

#apply prewitt edge detection
prewitt_x = cv.filter2D(image, -1, kernel_x)
prewitt_y = cv.filter2D(image, -1, kernel_y)

# Compute magnitude of gradients 
prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))

# Manual normalization to 0-255 range
prewitt_min = prewitt_combined.min()
prewitt_max = prewitt_combined.max()
if prewitt_max > prewitt_min:  # Check to avoid division by zero
    prewitt_combined = ((prewitt_combined - prewitt_min) * 255 / (prewitt_max - prewitt_min))

#display the original image and the prewitt edges
cv.imshow("Original", image)

# Convert to uint8 for display
cv.imshow("Prewitt Edges", np.uint8(prewitt_combined))

#wait for a key press and close all windows
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```

```python
#Robert edge detection

#load the image
image = cv.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg", cv.IMREAD_GRAYSCALE)

#perform guassian blur
image = cv.GaussianBlur(image, (5, 5), 0)

#define robert kernels
kernel_x = np.array([[1, 0], [0, -1]])

kernel_y = np.array([[0, 1], [-1, 0]])

#apply robert edge detection
robert_x = cv.filter2D(image, -1, kernel_x)
robert_y = cv.filter2D(image, -1, kernel_y)

#combine the gradient images
robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))

# Manual normalization to 0-255 range
robert_min = robert_combined.min()
robert_max = robert_combined.max()
if robert_max > robert_min:  # Check to avoid division by zero
    robert_combined = ((robert_combined - robert_min) * 255 / (robert_max - robert_min))

#display the original image and the robert edges
cv.imshow("Original", image)
cv.imshow("Robert Edges", np.uint8(robert_combined))

#wait for a key press and close all windows
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```

```python
# Laplacian edge detection

image = cv.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg", cv.IMREAD_GRAYSCALE)

image = cv.resize(image, (400, 400))

#perform Laplacian Edge Detection
laplacian = cv.Laplacian(image, cv.CV_64F)

#Convert the output to an 8-bit image
laplacian = cv.convertScaleAbs(laplacian)

#Display the original image and the Laplacian edges
cv.imshow("Original", image)
cv.imshow("Laplacian edges", laplacian)

#Wait for a key press and close all windows
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```

```python
# Canny edge detection

image = cv.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg", cv.IMREAD_GRAYSCALE)

image = cv.resize(image, (400, 400))

#Convert the output to an 8-bit image
edges = cv.Canny(image, 100, 200)

#Display the original image and the detected edges
cv.imshow("Original", image)
cv.imshow("edges", edges)

#Wait for a key press and close all windows
if cv.waitKey(0) & 0xff == 27:
    cv.destroyAllWindows()
```
