```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg")

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)

filtered_image = cv2.medianBlur(image, 15)

cv2.imwrite("../images/blurred_image.png",filtered_image )

plt.subplot(1,2,2)
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg")

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)

scaled_image = cv2.resize(image, None, fx =2, fy = 3)

cv2.imwrite("../images/scaled_image.png",scaled_image )

plt.subplot(1,2,2)
plt.title("Scaled Image")
plt.imshow(scaled_image)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("../images/Prof-Chris-Ogbechie-720x400-1-600x333.jpg")

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)

inverse_image = 255 - image

cv2.imwrite("../images/inverse_image.png",inverse_image )

plt.subplot(1,2,2)
plt.title("Inverse")
plt.imshow(inverse_image)
plt.show()
```
