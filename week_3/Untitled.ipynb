```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread("../images/Darlington-Agbholor-sst-300x225@2x.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

rows, cols, _ = img_rgb.shape

#plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img_rgb)

M = np.float32([[1,0,100], [0,1,50]])
transl = cv.warpAffine(img_rgb, M, (cols, rows))

#plot the translated image
plt.subplot(1,2,2)
plt.title("Translated Image")
plt.imshow(transl)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img = cv.imread("../images/Darlington-Agbholor-sst-300x225@2x.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

rows, cols, _ = img_rgb.shape

#plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img_rgb)

rows, cols, _ = img_rgb.shape
M = np.float32([[1,0,0], [0,-1,rows], [0,0,1]])
reflected_img = cv.warpPerspective(img_rgb, M, (int(cols), int(rows)))

#plot the original image
plt.subplot(1,2,2)
plt.title("Reflectd image")
plt.imshow(reflected_img)

cv.imwrite("reflected_out.jpg", reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image
image = cv.imread("../images/Darlington-Agbholor-sst-300x225@2x.jpg")

# Convert BGR to RGB for Matplotlib display
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Apply Gaussian Blur
gaussian = cv.GaussianBlur(image, (7, 7), 0)

# Apply Median Blur
median = cv.medianBlur(image, 5)

# Apply Bilateral Filter
bilateral = cv.bilateralFilter(image, 9, 75, 75)

# Convert all images to RGB for correct color display
gaussian_rgb = cv.cvtColor(gaussian, cv.COLOR_BGR2RGB)
median_rgb = cv.cvtColor(median, cv.COLOR_BGR2RGB)
bilateral_rgb = cv.cvtColor(bilateral, cv.COLOR_BGR2RGB)

# Create subplots
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

# Gaussian Blur
plt.subplot(2, 2, 2)
plt.imshow(gaussian_rgb)
plt.title("Gaussian Blur")
plt.axis("off")

# Median Blur
plt.subplot(2, 2, 3)
plt.imshow(median_rgb)
plt.title("Median Blur")
plt.axis("off")

# Bilateral Filter
plt.subplot(2, 2, 4)
plt.imshow(bilateral_rgb)
plt.title("Bilateral Filter")
plt.axis("off")

# Show all images
plt.tight_layout()
plt.show()
```
